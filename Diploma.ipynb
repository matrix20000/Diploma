# С использованием t-test статистики проверить взаимосвязи между несколькими переменными

Те кто работает меньше 150 часов, имеет такой же уровень удовлетворенности, что и те, кто работает более 200 часов

```python
hour_150 = data[data['average_montly_hours'] <= 150]['satisfaction_level']
hour_200 = data[data['average_montly_hours'] >= 200]['satisfaction_level']
```

```python
stats.shapiro(hour_150)
# ТК выборки не нормально распределены мы применяем U-тест 
```

```python
stat, p = stats.mannwhitneyu(hour_150, hour_200)
stat, p 
# p меньше 0.05, значит что мы можем принять нулевую гипотезу об отсутствии различий 
```

Сотрудники получающие низкую зарплату работают столько же, сколько и сотрудники с высокой зарплатой 

```python
def salary_(salary):
    if salary == 'low':
        return 1
    if salary == 'medium':
        return 2
    if salary == 'high':
        return 3

data.loc[:, 'salary_level'] = data['salary'].apply(salary_)
```

```python
time_sal_jun = data[data['salary_level'] == 1]['time_spend_company']
time_sal_sin= data[data['salary_level'] == 3]['time_spend_company']
```

```python
stats.shapiro(time_sal_jun)
# Хоть выборки не распределены нормально, но у них нет явных "горбов" на гистограмме, 
# поэтому попробум стандартный т-тест
```

```python
statistics.variance(time_sal_jun), statistics.variance(time_sal_sin)
# Дисперсия отличается, значит применим модификатор Уэлча
```

```python
stats.ttest_ind(time_sal_jun, time_sal_sin, equal_var=False)
# нулевая гипотеза также подтверждается визуально на графиках гистограмы, 
# нет различий между опытом и зп
```
